Every year, B Lab highlights the Certified B Corporations that go above and beyond in their pursuit of positive impact. All B Corporations take the B Impact Assessment and meet verified score requirements to become certified. In the assessment, companies are evaluated on their impact on the environment, their workers, their customers, and their communities; they also all meet the highest standards of accountability and transparency. Companies featured in the "Best for the World" list score in the top 10% of the B Corp community in each of seven categories—they are the top performers in an already exceptional group of businesses.

Research Questions: 
1. Which companies are the "Best for the World"? 
2. Which countries have the most Certified B Corporations? 
3. What are the industry categories with the best scores? 

Use a minimum of three plots and relevant statistics to back your choices. Think about which data is most relevant to answering the research question. Note: use `encoding = "ISO-8859-1"` when reading the CSV.

Data: https://static.junilearning.com/python_data_science/b_corp_impact.csv


```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

bcorpDF = pd.read_csv("https://static.junilearning.com/python_data_science/b_corp_impact.csv", encoding = "ISO-8859-1")
display(bcorpDF)
```

Which columns will we need to address the questions?

Which companies are the "Best for the World"? We need to find the top 10%.

Which countries have the most Certified B Corporations? We need the "Operating Country" column.

What are the industry categories with the best scores? We need the "Industry Category" column as well as the seven measurement columns near the right end of the df.

```
#Columns 14 through 54
newDF = bcorpDF.iloc[:,15:55]
display(newDF)
```

```
newDF.info()
```

```
newDF.isnull().sum()
```

```
overall = newDF.sum(axis=1,skipna=True)
print(overall.describe())
```

```
newDF["Overall Score"] = overall
display(newDF)
```

```
newDF["Company"] = bcorpDF["Name"]
display(newDF)
```

```
orderedDF = newDF.sort_values(by="Overall Score",ascending=False,ignore_index=True)
display(orderedDF)
```

```
top10 = orderedDF.iloc[0:10,:]
display(top10)
```

```
#Top 10 best companies
plt.xticks(rotation=90)
sns.barplot(y="Overall Score",x="Company",data=top10)
```

```
plt.figure().set_size_inches(18,9)
sns.countplot(x="Operating Country ",data=bcorpDF)
plt.xticks(rotation=90)
```

```
#Top 10 best industry categories
newDF["Industry Category"] = bcorpDF["Industry Category"]
display(newDF.sort_values(by = "Overall Score"))
plt.figure().set_size_inches(18,9)
plt.xticks(rotation=90)
sns.boxplot(y="Overall Score",x="Industry Category",data=newDF)
```

Which companies are the "Best for the World"? 

The "Best for the World" companies are: BioCarbon Partners, Eco2Librium, Dr. Bronners, South Mountain Company Inc., Fozenda Da Toca Organics, General Bio, Conscious Company Magazine, Impact Makers Inc, Beneficial State Bank, Ming Young Enterprise Co Ltd.


Which countries have the most Certified B Corporations? We need the "Operating Country"

The countries that had the most Certified B Corporations are: United States, Canada, Australia, and United Kingdom.

What are the industry categories with the best scores?

The top industry categories are: Energy (Heating & Cooling), Agriculture (Forestry & Fishing), and Energy.
